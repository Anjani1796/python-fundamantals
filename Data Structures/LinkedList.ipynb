```python
# Linked lists are the sequences of nodes pointing to each other nodes. Node has a value and link to other node. Its not 
# like array where insertion at a particular index is very expensive, so in  linked list the insertion  is easy cause we just 
# have to change the element previous element(set address in next part to new element) and new element(it should have the 
# next addressvalue set tot he previous element at this position) node values.
# Linked list traversal is also called link hopping or pointer hopping.
```

```python
# in case of arrays, if insertion has to be done it is expensive because it will swap other values to next subsequent addresses
# and in case there is no space for new element to be inserted then in that case it will take a new array of size = x + 2*x, 
# if x was earlier the size of array. This is a drawback in case of the arrays
# arrays store values in continuous memory location, but in linked lists element are stored at random memory location which 
# are linked to each other 

# Time complexities
# insert/delete at beginning at O(1)
# insert/delete at end at O(n)
# linked list traversal - O(n)
# Accessing element by value - O(n)

# linked lists have two main disadvantages:
# 1. no need to preaallocate space
# 2. insertion is easier
```

```python
# in doubly linked lists, the backward traversal becomes easy, coz we store the previous value also

# array vs Linked list time complexity, first complexity is of arrays and other for linked list
# indexing - O(1) O(n)
# insert/delete at start - O(n) O(1)
# insert/delete at end - O(1) O(n)
# insert in middle - O(n) O(n)

# in arrays the traversal is fast if you know the index and in linked list you have to travel the whole linked list either
# forward or backward
```

```python
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next
        
class LinkedList:
    def __init__(self):
        self.head = None 
    def insertAtBeginning(self, data):
        node = Node(data, self.head)
        self.head = node
    def print(self):
        if self.head == None:
            print("Linked list is empty")
            
        itr = self.head 
        llstr = ''
        while itr:
            llstr += str(itr.data) + '-->'
            itr = itr.next
        print(llstr)
        
    def print_palindrome(self):
        if self.head == None:
            print("Linked list is empty")
            
        itr = self.head 
        llstr = ''
        while itr:
            llstr += str(itr.data)
            itr = itr.next
        print(llstr)
        if llstr == llstr[::-1]:
            print("list is palindrome")
        
    def insertAtEnd(self, data):
        if self.head is None:
            self.head = Node(data, None)
            return
        
        itr = self.head
        while itr.next:
            itr = itr.next
        itr.next = Node(data, None)
    
    def insertValues(self, data_list):
        self.head = None
        
        for data in data_list:
            self.insertAtEnd(data)
    
    def getLength(self):
        count = 0
        itr = self.head
        while itr:
            count += 1
            itr = itr.next
        return count
    
    def removeAt(self, index):
        if index<0 or index >= self.getLength():
            raise Exception('Invalid Index')
        if index == 0:
            self.head = self.head.next
            return
        
        count = 0
        itr = self.head
        
        while itr:
            if count == index - 1:
                itr.next = itr.next.next
                break
            itr = itr.next
            count += 1
            
    def insertAtPosition(self, index, data):
        if index<0 or index >= self.getLength():
            raise Exception('Invalid Index')
        if index == 0:
            self.insertAtBeginning(data)
            return
        count = 0
        itr = self.head
        while itr:
            if count == index - 1:
                node = Node(data, itr.next)
                itr.next = node
                break
            itr = itr.next
            count += 1
                
            
if __name__ == '__main__':
    ll = LinkedList()
    ll.insertValues(['1', '2', '2', '1'])
    ll.print()
    ll.insertAtBeginning(5)
    ll.insertAtBeginning(89)
    
    ll.print()
    
#     ll.insertAtEnd(75)

#     ll.insertAtEnd(79)
#     ll.insertAtEnd(1)
#     ll.insertAtEnd(9876)

#     ll.insertValues(['banana', 'mango', 'grapes', 'orange'])
#     ll.insertValues(['1', '2', '2', '1'])
#     ll.print()
    #89-->5-->75-->
    
#     print("Length:", ll.getLength())
    
#     ll.removeAt(2) #remove the element present at index 2
#     ll.removeAt(20) #remove the element present at index 20 which is not present, it throws invalid index error
#     ll.print()
#     banana-->mango-->grapes--> orange-->
#     banana-->mango--> orange-->


#     insert at given index
#     ll.insertAtPosition(0, 'figs')
#     ll.insertAtPosition(2, 'jackfruit')
#     ll.print_palindrome()
    
    
```
